```python
df_movies=df_movies[['genres','movie_id','keywords','title','overview','crew','cast']]
```

```python
df_movies.dropna(inplace=True)
```

```python
df_movies['genres'][0]
```

```python
import ast 
def convert(obj):
     L=[]
     for i in ast.literal_eval(obj) :
          L.append(i['name'])
     return L
```

```python
df_movies['genres'] = df_movies['genres'].apply(convert)
```

```python
df_movies['cast'][0]
```

```python
import ast 
def Cast_Convert(obj):
     L=[]
     breaker = 0
     for i in ast.literal_eval(obj) :
          if breaker != 3:
               L.append(i['name'])
               breaker+=1
          else:
               break
     return L
```

```python
df_movies['cast'] = df_movies['cast'].apply(Cast_Convert)
```

```python
df_movies['cast']
```

```python
df_movies['crew'][0]
```

```python
df_movies['keywords'] = df_movies['keywords'].apply(convert)
```

```python
df_movies.head(2)
```

```python
def fetch_director(obj):
     L=[]
     for i in ast.literal_eval(obj) :
          if i['job'] == 'Director' :
               L.append(i['name'])
               break
     return L
```

```python
df_movies['crew'] = df_movies['crew'].apply(fetch_director)
```
