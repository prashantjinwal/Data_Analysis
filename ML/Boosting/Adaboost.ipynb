```python
import pandas as pd
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.ensemble import AdaBoostClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
data = load_breast_cancer()

X = data.data
Y= data.target
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, Y, test_size=0.2, random_state=42
)
```

```python
model = AdaBoostClassifier(n_estimators=50, learning_rate=1.0, random_state=42)
```

```python
model.fit(X_train, y_train)

y_pred = model.predict(X_test)
```

```python
print("Accuracy:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
```

## Using a Different Base Estimator

By default, AdaBoost uses a decision stump (a decision tree with max_depth=1) as the weak learner.

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
base_model = DecisionTreeClassifier(max_depth=2,random_state=42)
```

```python
model2 = AdaBoostClassifier(estimator=base_model,n_estimators=100, learning_rate=1.0, random_state=42)
```

```python
model2.fit(X_train, y_train)

y_pred = model2.predict(X_test)
```

```python
print("Accuracy:", accuracy_score(y_test, y_pred))
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("MAE :", mae)
print("MSE :", mse)
print("R² Score :", r2)
```
