### Type of cross validation 
1) LOOCV (leave one out cross validation)
2) Hold out cross validation
3) K-Fold cross validation 
4) Stratified K-Fold cross validation
5) Time Series cross validation


# 1. Hold-Out Cross Validation

## Definition
The dataset is divided into two parts:
- Training Set (e.g., 80%)
- Testing Set (e.g., 20%)

The model is trained on the training set and evaluated on the testing set.

## Advantages
- Simple and fast
- Suitable for large datasets

## Disadvantages
- Performance depends on the random split
- Less reliable for small datasets

## Example
80% Train → 20% Test

# 2. K-Fold Cross Validation

## Definition
The dataset is divided into **K equal folds**. The model is trained and tested K times.

For each iteration:
- 1 fold is used for testing
- Remaining K−1 folds are used for training

Final performance = Average of all K results.

## Advantages
- Better performance estimation
- Efficient use of data

## Disadvantages
- More computationally expensive than Hold-Out

## Example
For K = 5:

- Fold 1 → Test, Others → Train
- Fold 2 → Test, Others → Train
- Fold 3 → Test, Others → Train
- Fold 4 → Test, Others → Train
- Fold 5 → Test, Others → Train

# 3. Stratified K-Fold Cross Validation

## Definition
A variation of K-Fold where each fold preserves the same class distribution as the original dataset.

## Why Use It?
Useful for imbalanced classification datasets.

## Advantages
- Maintains class proportions
- More reliable evaluation

## Disadvantages
- Slightly more complex than standard K-Fold

## Example
Original Dataset:
- Class A = 80%
- Class B = 20%

Each fold will also contain:
- Class A ≈ 80%
- Class B ≈ 20%

# 4. Leave-One-Out Cross Validation (LOOCV)

## Definition
A special case of K-Fold where:

K = Number of Samples (N)

For each iteration:
- Train on N−1 samples
- Test on 1 sample

## Advantages
- Maximum data used for training
- Low bias

## Disadvantages
- Computationally expensive
- High variance

## Example
Dataset = 100 samples

- Train on 99, Test on 1
- Repeat 100 times

# 5. Time Series Cross Validation

## Definition
Used when data has a time order. Future data must not be used to predict past data.

The training set grows over time.

## Advantages
- Prevents data leakage
- Suitable for forecasting tasks

## Disadvantages
- Less training data in early iterations

## Example

Iteration 1:
- Train: [1,2,3]
- Test: [4]

Iteration 2:
- Train: [1,2,3,4]
- Test: [5]

Iteration 3:
- Train: [1,2,3,4,5]
- Test: [6]

# Quick Comparison

| Method | Best For | Speed |
|----------|----------|--------|
| Hold-Out | Large datasets | Fast |
| K-Fold | General ML tasks | Medium |
| Stratified K-Fold | Imbalanced classification | Medium |
| LOOCV | Small datasets | Slow |
| Time Series CV | Forecasting problems | Medium |
