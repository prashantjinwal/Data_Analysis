```python
import matplotlib.pyplot as plt

x = [1,2,8,5]
y = [2,5,8,4]
plt.plot(x,y)
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("Basic line plot")
plt.show()
```

```python
#customized line plot

x = [1,2,3,5]
y = [2,5,2,4]
plt.plot(x,y, color="red", linestyle="--", marker="o",linewidth="3",markersize="8")
plt.grid(True)
```

```python
#multiple plot 
x = [1,2,8,5]
y1 = [2,5,8,4]
y2 = [1,3,4,6]

plt.figure(figsize=(9,5))
plt.subplot(2,2,1)
plt.plot(x,y1, color="green")
plt.title("Plot 1")

plt.subplot(2,2,2)
plt.plot(x,y2, color="red")
plt.title("Plot 2")

plt.subplot(2,2,3)
plt.plot(y1,x, color="yellow")
plt.title("Plot 3")

plt.show()
```

```python
#bar plot 
categories=['A','B','C','D']
value=[2,5,4,6]

plt.bar(categories,value,color="purple")
plt.xlabel("categories")
plt.ylabel("value")
plt.title("bar plot")
plt.show()
```

```python
# HIST plot
data=[1,2,2,2,2,2,4,4,7,5,7,6,6,6,3]
plt.hist(data,bins=7,color='orange',edgecolor='black')
```

```python
#Scatter plot

x=[1,2,4,6,3]
y=[2,3,4,5,6]
plt.scatter(x,y,color='blue',marker='x')
```

```python
### pie plot 
labels=['A','B','C','D']
size=[20,50,40,60]
color=['gold', 'yellow', 'blue', 'red']
explode=(0.2,0,0,0)

plt.pie(size,labels=labels)

```
