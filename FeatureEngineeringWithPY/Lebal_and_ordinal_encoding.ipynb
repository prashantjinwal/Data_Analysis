## lebal and ordinal encoding

```python
import pandas as pd 
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.DataFrame({
    'color': ['red', 'blue','green', 'green' ,'red', 'blue']
})
```

```python
df_encoder = LabelEncoder()
```

```python
df_encoder.fit_transform(df[['color']])
```

```python
df_encoder.transform([['red']])
```

## ordinal encoding

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
size_df = pd.DataFrame({
     'size' : ['small', 'medium', 'large', 'small','large','large']
}) 
```

```python
encoded = OrdinalEncoder(categories=[['small', 'medium', 'large']])
```

```python
encoded.fit_transform(size_df[['size']])
```
