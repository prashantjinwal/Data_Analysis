## 5 number summary


Maximum, minimum, Q1, Q2, Median, Median 

```python
import numpy as np

lst_age = [23,44,67,85,34,99,75,43,45,89,95,56,77,63,76,23,42]
maximum,Q1,Median,Q3,minimum = np.quantile(lst_age,[0, 0.25, 0.50, 0.75, 1.0])
```

```python
maximum,Q1,Median,Q3,minimum
```

```python
IQR = Q3 - Q1
print(IQR)
```

```python
lower_fence=Q1 - 1.5*(IQR)
higher_fence=Q3 + 1.5*(IQR)
```

```python
print(lower_fence)
print(higher_fence)
```
