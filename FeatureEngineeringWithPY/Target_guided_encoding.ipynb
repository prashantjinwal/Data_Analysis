```python
import pandas as pd
```

```python
df = pd.DataFrame({
     'city' : ['Delhi', 'Up', 'Delhi', 'Mumbai', 'Up', 'Delhi'],
     'price' : [233,455,772,334,323,443]
})
```

```python
df
```

```python
mean_price = df.groupby('city')['price'].mean().to_dict()
```

```python
mean_price
```

```python
df['encoded_'] = df['city'].map(mean_price)
df
```

```python
import seaborn as sns
df_tip = sns.load_dataset('tips')
print(df_tip)
```

```python
mean_tip = df_tip.groupby('time', observed=True)['total_bill'].mean().to_dict()
```

```python
mean_tip
```

```python
df_tip['encoded_timetip'] = df_tip['time'].map(mean_tip)
df_tip
```
