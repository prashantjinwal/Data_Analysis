## SMOTE (Synthetic Minority Oversampling Technique)
Itâ€™s a data preprocessing method used in imbalanced datasets, where one class (usually the minority class, like fraud cases, disease-positive patients, etc.) has far fewer samples than the majority class.

```python
from sklearn.datasets import make_classification
X,Y = make_classification(n_samples=1000, n_features=2, n_redundant=0, n_clusters_per_class=1, weights=[0.90], random_state=12)
```

```python
import pandas as pd
df1 = pd.DataFrame(X ,columns=['f1','f2'])
df2 = pd.DataFrame(Y ,columns=['target'])
final_df = pd.concat([df1,df2],axis=1)
final_df.head()
```

```python
final_df['target'].value_counts()
```

```python
import matplotlib.pyplot as plt
plt.scatter(final_df['f1'], final_df['f2'], c=final_df['target'])
```

```python
from imblearn.over_sampling import SMOTE
```

```python
oversampling = SMOTE()
X,Y = oversampling.fit_resample(final_df[['f1','f2']],final_df['target'])
```
